```python
import pandas as pd
import numpy as np
import tabulate
```

## AIRBNB BUSINESS MANAGEMENT

```python
def AirBnB_apartment_management(filename):
# Loading the AirBnB file into python
        dataset= pd.read_excel(filename,dtype={'id':str})
        df = dataset.copy()
        df.describe() # A quick statistic about every variable of the file
        df.info()    # information about the number of rows and data types
        df.isnull().sum() # To check for null or NAN values if any

# replacing missing values in the NAME house_rules and host name columns respectively
# it should be noted that if a column is to be used for predictive analysis and that column missing value is above 30%: reject the column
        df.fillna({'NAME': 'AirBnB facility name not written'},inplace=True)

        df.fillna({'house_rules': 'Rules not written,but your discretion is highly welcome'},inplace=True)

        df.fillna({'host name': 'Greg'},inplace=True)

# Droping duplicates rows
        df = df.drop_duplicates()

# Shape after droping duplicate rows
        df.shape
        for column in df.columns:
                if pd.api.types.is_numeric_dtype(column):
                        df.fillna({column:df[column].mean()},inplace = True)
                        Q1 = df[column].quantile(0.25)
                        Q3 = df[column].quantile(0.75)
                        IQR = Q3 - Q1
                        lower = Q1 - 1.5 * IQR
                        upper = Q3 + 1.5 * IQR
                        mean = df[column].mean()
                        df[column] = df[column].apply(lambda x: mean if x < lower or x > upper else x)
                else:
                        df.fillna({column:df[column].mode()[0] if df[column].mode()[0] is not None else "Unknown"},inplace = True)
        return df
      
```

```python
class AirBnB_renters:
            def __init__(self,df):
                self.home = df
        #AirBnB Accomodations facility by room type and year of construction    
            def room (self):
                roomtype = input("Enter Room Type").strip().lower()
                year_of_construction = int(input("Enter Year of Construction"))
                room_details = []
                
                for _,row in self.home.iterrows():
                     room =row['room type'].strip().lower()
                     room_id = row['id']
                     year = row['Construction year']
                     price = row['price']
                     minimum_nights = row['minimum nights']
                     room_reviews = row['number of reviews']
                     service_fee = row['service fee']

                     if roomtype ==  room and year_of_construction == year:
                          room_details.append({'Selected Room':room,'Room ID':room_id,'Price':price,
                                               'Nights Allowed':minimum_nights,
                                               'Reviews': room_reviews,
                                               'Service Charge':service_fee,
                                               'year constructed':year})
                
                if room_details:
                    room_details_df =pd.DataFrame(room_details)
                    print(f"Grace AirBnB Limited\nNumber of rooms:{len(room_details_df)}| Room type:{roomtype}| Construction year:{year_of_construction}\n"
                          f"{tabulate.tabulate(room_details_df.set_index('Selected Room'),headers=room_details_df.keys(),tablefmt='heavy_grid')}")
                    return self.home,room_details_df
                else:
                        print(f'Room combination\nRoom Type:{roomtype} Construction year:{year_of_construction} Not Available')
                        return self.home, pd.DataFrame()
#verifying if a room have been checked in using the ID
            def confirm_checkin(self,df,df1):
                self.allrooms = df
                self.selected_room = df1
                checkin_info = []
                if all(bothdf is not None and not bothdf.empty for bothdf in [self.allrooms,self.selected_room]):
                    
                    try:
                        
                        checkedinID= (input("WELCOME to Grace AirBnB renters\n"
                                                "Convenience at its peak!\n"
                                                "Enter Checkin ID"))
                        if checkedinID not in self.selected_room['Room ID'].values:
                            print(f"Check-in ID {checkedinID} not found in selected room list.")
                            return

                        for _,row in self.allrooms.iterrows():
                            if row['id'] == checkedinID:
                                TypeA = row['room type']
                                name = row['NAME']
                                hostname = row['host name']
                                price = row['price']
                                service_fee = row['service fee']
                                cancellation_p = row['cancellation_policy']
                                neighboor = row['neighbourhood']
                                datecheckedin = row['last review']
                            
                            
                                print(f"Room Reservation ID:{checkedinID} CONFIRMED")
                                confirmation_note = input(f"Proceed to view Reservation details?\nEnter (Yes) to confirm (No) to terminate")
                                if confirmation_note.strip().lower() == 'yes':
                                    checkin_info.append({'Type of Room': TypeA ,
                                                        'Facility Description': name,
                                                        'Customer Name':hostname,'Price':price,
                                                        'Service_fee':service_fee,
                                                        'Cancellation_policy': cancellation_p,
                                                        'Environs':neighboor,
                                                        'Datecheckedin':datecheckedin})
                                    
                                elif confirmation_note.strip().lower() =='no':
                                    print("Session terminated")
                                else:
                                    print('Invalid response')
                    
                        
                        if checkin_info:
                            details = pd.DataFrame(checkin_info)
                            print(f'Client checked-in details\n{tabulate.tabulate(details.set_index('Type of Room').head(101),headers = details.columns.tolist(),tablefmt='heavy_grid')}')
                            
                        else:
                            print(f"No matching details for check-in ID: {checkedinID} found.")
                    except ValueError:
                        print("Invalid input. Please enter a valid ID.")
                else:
                        print("One or both dataframe is missing or empty, Cannot Proceed")
```

```python
AirbnbData = AirBnB_apartment_management("cleanedAirBnB.xlsx")
Grace_AirBnB = AirBnB_renters(AirbnbData)
Allrooms,Roomtypeselected = Grace_AirBnB.room()
verify_checkin = Grace_AirBnB.confirm_checkin(Allrooms,Roomtypeselected)
```
